## Function to resize and restore the images

```python

def resizeImage(images):
    img_rows, img_cols = 100,100
    im = Image.open(subdir + "\\" + images)
    img = im.resize((img_rows, img_cols))    
    gray = img.convert('L')
    gray.save(_foldert+"/"+images,"PNG")
```

# Traverse the complete image directory structure 

```python
counter = 1
for subdir, dirs, files in os.walk(_folder):
    #print("\n"+subdir)
    for images in files:
        if counter == 11:
            counter = 2
            break
        else:
            print("-->" + subdir + "\\" + images)
            resizeImage(images)
            counter += 1
            
```
